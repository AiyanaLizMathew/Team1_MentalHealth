```python
# A funtion to find a weighted variable based on various factors as the Mental Health Value
def CalculatedVariable(DF):
    
    # For each record in the dat
    for index,row in DF.iterrows():
        
        # Intializing the variables
        CalcVar=0
        TotalWght=0
        
        # Priority 1
        
        # For General Health 
        if row["General_Health"]=="Poor":
            CalcVar+=10
        else:
            CalcVar+=0
        
        # For Smokers
        if row["Smoke_Frequency"]=="Smoker":
            CalcVar+=10
        else:
            CalcVar+=0
        
        # For Alcoholics
        if row["Alcohol_Frequency"]=="Alcoholic":
            CalcVar+=10
        else:
            CalcVar+=0

        # For Sleeping disorders
        if row["Sleep_Hours"]<5 and row["Sleep_Hours"]>10:
            CalcVar+=10
        else:
            CalcVar+=0
        
        # For Obesity, Overweight and Underweight
        if row["BMI"]=="Obese":
            CalcVar+=10
        elif row["BMI"]=="Overweight":
            CalcVar+=8
        elif row["BMI"]=="Underweight":
            CalcVar+=8
        else:
            CalcVar+=0

        # For Physical Activity
        if row["Physical_Activity"]=="N":
            CalcVar+=10
        else:
            CalcVar+=0

        # For Chronic Disease
        # As there is already a existing value in the score of 10 adding that value to the Calculated Variable
        CalcVar+=row["Chronic_Disease"]

        # For Pregnancy
        if row["Pregnant"]=="Y":
            CalcVar+=10
        else:
            CalcVar+=0
    
    
        # Priority 2
        # For Income
        if row["Income"]=="<15000":
            CalcVar+=5
        elif row["Income"]=="15000-<25000":
            CalcVar+=4
        elif row["Income"]=="25000-<35000":
            CalcVar+=3
        elif row["Income"]=="35000-<50000":
            CalcVar+=2
        elif row["Income"]=="50000+":
            CalcVar+=1

        # Priority 3
        # For Emotional Support Availability
        if row["EmotionalSupport_Availability"]==1:
            CalcVar+=0.5
        elif row["EmotionalSupport_Availability"]==2:
            CalcVar+=1
        elif row["EmotionalSupport_Availability"]==3:
            CalcVar+=1.5
        elif row["EmotionalSupport_Availability"]==4:
            CalcVar+=2
        elif row["EmotionalSupport_Availability"]==5:
            CalcVar+=2.5
        # For Life Satisfaction Level
        if row["LifeSatisfaction_Level"]=="Not-Satisfied":
            CalcVar+=2.5
        else:
            CalcVar+=0
       
        # If the person has reported Chronic Depression
        if row["Chronic_Depression"]=="Y":
            chronic_dep=10
            # Converting the stress days(1-30) to a score of 10
            stress = row["Stress_Days"]*(1/3)
             # Converting the depressed days(1-30) to a score of 10
            depress=row["Depressed_Days"]*(1/3)
        else:
            # Converting the stress days(1-30) to a score of 10
            stress=row["Stress_Days"]*(1/3)
            # Converting the depressed days(1-30) to a score of 10
            depress=row["Depressed_Days"]*(1/3)
            chronic_dep=1
            
        # Calculating the average mental factor value
        avg_depr=(stress+depress+chronic_dep)*(2/3)
        
        # Checking if male or female and based on that converting the score to 100
        if row["Gender"]=="M":
            TotalWght=(CalcVar+avg_depr)*100/100
        else:
            TotalWght=(CalcVar+avg_depr)*100/110
    
        DF.set_value(index,"Calc_MentalHealth_Value",round(TotalWght,2))   

```
